## What is Synthetic Biomedical Data?

**Synthetic biomedical data** is artificially generated data that mimics the statistical properties of real patient data
without containing any real patient information.

---

## Why use synthetic biomedical data?

1. **Privacy-preserving research** — share and analyze data without violating privacy laws.
2. **Benchmarking algorithms** — test feature selection, classification, and other ML methods.
3. **Educational purposes** — teach students without using real patient data.
4. **Prototyping and software testing** — develop pipelines before deploying on sensitive data.

```python
# --- Cell 2: Imports ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification

# Set style for plots
sns.set_theme(style="whitegrid")
```

```python
# --- Cell 3: Generate synthetic biomedical-like data ---
# Simulate 100 patients, 5 biomedical features, binary disease status
X, y = make_classification(
    n_samples=100,
    n_features=5,
    n_informative=3,
    n_redundant=0,
    n_classes=2,
    random_state=42
)

# Convert to DataFrame
feature_names = [f"biomarker_{i+1}" for i in range(X.shape[1])]
df = pd.DataFrame(X, columns=feature_names)
df["disease_status"] = y  # 0 = healthy, 1 = diseased

df.head()
```

```python
# --- Cell 4: Basic data exploration ---
print("Dataset shape:", df.shape)
print("\nClass distribution:")
print(df["disease_status"].value_counts())

# Summary statistics
df.describe()
```

```python
# --- Cell 5: Visualize feature distributions ---
plt.figure(figsize=(10, 6))
for i, col in enumerate(feature_names[:3], 1):
    plt.subplot(1, 3, i)
    sns.histplot(data=df, x=col, hue="disease_status", kde=True, element="step")
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# --- Cell 6: Simple scatter plot between two biomarkers ---
plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=df,
    x="biomarker_1",
    y="biomarker_2",
    hue="disease_status",
    palette="Set1",
    s=80
)
plt.title("Biomarker 1 vs Biomarker 2")
plt.show()
```

```python
# --- Cell 6: Simple scatter plot between two biomarkers ---
plt.figure(figsize=(6, 6))
sns.scatterplot(
    data=df,
    x="biomarker_1",
    y="biomarker_2",
    hue="disease_status",
    palette="Set1",
    s=80
)
plt.title("Biomarker 1 vs Biomarker 2")
plt.show()
```
